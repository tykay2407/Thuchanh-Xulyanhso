```python
#cau1

import numpy as np
import imageio.v2 as iio
import scipy.ndimage as nd
import matplotlib.pylab as plt

# Bước 1: Đọc ảnh từ thư mục 'exercise'
image = iio.imread('exercise/colorful-ripe-tropical-fruits.jpg')

# Bước 2: Cắt đúng quả kiwi như bạn gửi mẫu (gần trái thơm và quả dừa)
kiwi = image[870:930, 460:520]  # Chính xác quả kiwi tròn

# Bước 3: Lưu ảnh kiwi gốc
iio.imsave('exercise/kiwi_original.jpg', kiwi.astype(np.uint8))

# Bước 4: Tịnh tiến kiwi sang phải 30 pixels
kiwi_shifted = nd.shift(kiwi, shift=(0, 30, 0))  # dịch sang phải

# Bước 5: Lưu ảnh kiwi đã dịch
iio.imsave('exercise/kiwi_shifted.jpg', kiwi_shifted.astype(np.uint8))

# Bước 6: Hiển thị cả hai ảnh
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].imshow(kiwi.astype(np.uint8))
axs[0].set_title("Kiwi gốc (đúng quả bạn chọn)")
axs[0].axis('off')

axs[1].imshow(kiwi_shifted.astype(np.uint8))
axs[1].set_title("Kiwi đã dịch phải 30px")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#cau 2

import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
import scipy.ndimage as nd

# Đọc ảnh từ thư mục 'exercise'
image = iio.imread('exercise/colorful-ripe-tropical-fruits.jpg')

# --- Cắt quả đu đủ ---
papaya = image[670:770, 190:270]

# --- Cắt quả dưa hấu (đúng quả to, lát đỏ như bạn gửi) ---
watermelon = image[580:760, 1550:1800]

# --- Đổi màu: dùng đảo màu (inversion)
papaya_inverted = 255 - papaya
watermelon_inverted = 255 - watermelon

# --- Lưu ảnh riêng ---
iio.imsave('exercise/papaya_inverted.jpg', papaya_inverted.astype(np.uint8))
iio.imsave('exercise/watermelon_inverted.jpg', watermelon_inverted.astype(np.uint8))

# --- Hiển thị kiểm tra ---
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].imshow(papaya.astype(np.uint8))
axs[0, 0].set_title("Đu đủ gốc")
axs[0, 0].axis('off')

axs[0, 1].imshow(papaya_inverted.astype(np.uint8))
axs[0, 1].set_title("Đu đủ đổi màu")
axs[0, 1].axis('off')

axs[1, 0].imshow(watermelon.astype(np.uint8))
axs[1, 0].set_title("Dưa hấu gốc")
axs[1, 0].axis('off')

axs[1, 1].imshow(watermelon_inverted.astype(np.uint8))
axs[1, 1].set_title("Dưa hấu đổi màu")
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()

```

```python
#cau 3

import numpy as np
import imageio.v2 as iio
import scipy.ndimage as nd
import matplotlib.pylab as plt

# Đọc ảnh gốc
image = iio.imread('exercise/quang_ninh.jpg')

# Cắt ngọn núi (vẫn giữ nguyên như trước)
mountain = image[100:400, 470:670]

# Cắt vùng con thuyền mở rộng để thấy rõ hơn
boat = image[450:540, 400:540]

# Xoay cả hai đối tượng
mountain_rotated = nd.rotate(mountain, 45, reshape=True)
boat_rotated = nd.rotate(boat, 45, reshape=True)

# Lưu vào thư mục exercise
iio.imsave('exercise/mountain_rotated.jpg', mountain_rotated.astype(np.uint8))
iio.imsave('exercise/boat_rotated.jpg', boat_rotated.astype(np.uint8))

# Hiển thị kiểm tra
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(mountain_rotated.astype(np.uint8))
axs[0].set_title("Ngọn núi xoay 45°")
axs[0].axis('off')

axs[1].imshow(boat_rotated.astype(np.uint8))
axs[1].set_title("Con thuyền xoay 45° (mở rộng)")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#cau4

import numpy as np
import imageio.v2 as iio
import scipy.ndimage as nd
import matplotlib.pylab as plt

# Đọc ảnh pagoda từ thư mục exercise
image = iio.imread('exercise/pagoda.jpg')

# Cắt ngôi chùa ở trung tâm ảnh
pagoda = image[100:300, 100:300]

# Phóng to gấp 5 lần
pagoda_scaled = nd.zoom(pagoda, zoom=(5, 5, 1))  # (height, width, channels)

# Lưu ảnh đã phóng to
iio.imsave('exercise/pagoda_scaled_5x.jpg', pagoda_scaled.astype(np.uint8))

# Hiển thị để kiểm tra
plt.imshow(pagoda_scaled.astype(np.uint8))
plt.title("Ngôi chùa phóng to 5 lần")
plt.axis('off')
plt.show()
```

```python
#cau5

import numpy as np
import imageio.v2 as iio
import scipy.ndimage as nd
import matplotlib.pyplot as plt

# Danh sách ảnh có sẵn
images = {
    "1": "exercise/colorful-ripe-tropical-fruits.jpg",
    "2": "exercise/quang_ninh.jpg",
    "3": "exercise/pagoda.jpg"
}

def show_coordinate_map(image):
    plt.imshow(image)
    plt.grid(True)
    plt.title("Coordinate Map")
    plt.show()

def apply_translation(image):
    dx = int(input("Nhập số pixel dịch theo chiều ngang (dx): "))
    dy = int(input("Nhập số pixel dịch theo chiều dọc (dy): "))
    shifted = nd.shift(image, shift=(dy, dx, 0))
    plt.imshow(shifted.astype(np.uint8))
    plt.title("Ảnh sau khi tịnh tiến")
    plt.axis('off')
    plt.show()

def apply_rotation(image):
    angle = float(input("Nhập góc xoay (độ): "))
    rotated = nd.rotate(image, angle, reshape=True)
    plt.imshow(rotated.astype(np.uint8))
    plt.title(f"Ảnh xoay {angle}°")
    plt.axis('off')
    plt.show()

def apply_zoom(image, scale):
    zoomed = nd.zoom(image, (scale, scale, 1))
    plt.imshow(zoomed.astype(np.uint8))
    title = "Ảnh phóng to" if scale > 1 else "Ảnh thu nhỏ"
    plt.title(f"{title} (x{scale})")
    plt.axis('off')
    plt.show()

# Menu chính
print("===== MENU BIẾN ĐỔI ẢNH =====")
print("T - Tịnh tiến ảnh")
print("X - Xoay ảnh")
print("P - Phóng to ảnh")
print("H - Thu nhỏ ảnh")
print("C - Hiển thị bản đồ tọa độ")
choice = input("Chọn thao tác (T/X/P/H/C): ").strip().upper()

print("\nChọn ảnh muốn áp dụng:")
for k, v in images.items():
    print(f"{k} - {v.split('/')[-1]}")
img_choice = input("Nhập số (1/2/3): ").strip()

# Đọc ảnh đã chọn
if img_choice in images:
    image = iio.imread(images[img_choice])
else:
    print("Ảnh không hợp lệ.")
    exit()

# Thực hiện thao tác tương ứng
if choice == "T":
    apply_translation(image)
elif choice == "X":
    apply_rotation(image)
elif choice == "P":
    apply_zoom(image, scale=2.0)
elif choice == "H":
    apply_zoom(image, scale=0.5)
elif choice == "C":
    show_coordinate_map(image)
else:
    print("Lựa chọn không hợp lệ.")
```
